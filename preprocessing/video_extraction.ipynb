```python
#library imports
import pandas as pd
import os
import cv2
```

```python
labels = pd.read_csv("/content/drive/MyDrive/accident_files/sim_dataset/labels.csv")
labels
```

```python
T_bone_labels = labels[labels["type"] == "single"]
T_bone_labels
```

```python
import pandas as pd
import os
import cv2

print("Original sideswipe_labels shape:", T_bone_labels.shape)
print("Weather type distribution:\n", T_bone_labels['weather'].value_counts())

# Determine the number of videos to pick per weather type to ensure equal proportion.
# Let's target sampling up to 10 videos per weather type, but not exceeding the available count.
sample_target_per_weather = 10

selected_videos = pd.DataFrame()
unique_weather_types = T_bone_labels['weather'].unique()

for weather_type in unique_weather_types:
    weather_group = T_bone_labels[T_bone_labels['weather'] == weather_type]
    num_available = len(weather_group)

    if num_available == 0:
        print(f"Warning: No videos found for weather type '{weather_type}'.")
        continue

    # Sample either the target number, or all available if fewer than target
    num_to_sample = min(sample_target_per_weather, num_available)

    # Use random_state for reproducibility of the sample
    sampled_group = weather_group.sample(n=num_to_sample, random_state=42)
    selected_videos = pd.concat([selected_videos, sampled_group])

print(f"\nSelected {len(selected_videos)} videos for frame extraction:")
print(selected_videos['weather'].value_counts())
display(selected_videos)
```

```python
import os

file_path = "/content/drive/MyDrive/accident_files/selected_videos.csv"

selected_videos.to_csv(file_path, mode='a', header=False, index=False)
print(f"Appended {len(selected_videos)} videos to {file_path}")
```

```python
base_video_dataset_path = '/content/drive/MyDrive/accident_files/sim_dataset/'
output_base_dir = '/content/drive/MyDrive/accident_files/random_extracted_frames/'
output_frames_dir = os.path.join(output_base_dir, 'frames')

os.makedirs(output_frames_dir, exist_ok=True)
print(f"\nOutput frames will be saved in: {output_frames_dir}")
```

```python
# Frame extraction loop
extracted_video_count = 0
total_frames_extracted = 0

for index, row in selected_videos.iterrows():
    relative_rgb_path = row['rgb_path'] # e.g., videos/head-on/Town05_head-on_night_34.mp4
    full_video_path = os.path.join(base_video_dataset_path, relative_rgb_path)

    # Get the base filename without path or extension
    # e.g., 'videos/head-on/Town05_head-on_night_34.mp4' -> 'Town05_head-on_night_34'
    base_video_name = os.path.splitext(os.path.basename(relative_rgb_path))[0]
    filename_prefix = base_video_name

    cap = cv2.VideoCapture(full_video_path)

    if not cap.isOpened():
        print(f"    Error: Could not open video {full_video_path}")
        continue

    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break # End of video

        # Name file as videofilename_frame_xxxx.jpg
        frame_filename = f"{filename_prefix}_frame_{frame_count:04d}.jpg" # Using 4 digits for frame number
        frame_output_path = os.path.join(output_frames_dir, frame_filename)

        cv2.imwrite(frame_output_path, frame)
        frame_count += 1

    cap.release()
    print(f"    Extracted {frame_count} frames from {relative_rgb_path}")
    extracted_video_count += 1
    total_frames_extracted += frame_count

print(f"\nCompleted frame extraction for {extracted_video_count} selected videos. Total frames: {total_frames_extracted}.")
```

```python
from sklearn.model_selection import train_test_split

base_video_dataset_path = '/content/drive/MyDrive/accident_files/sim_dataset/'
output_frames_dir = '/content/drive/MyDrive/accident_files/smaller_sampled_frames'
os.makedirs(output_frames_dir, exist_ok=True)

print(f"\nOutput frames will be saved in: {output_frames_dir}")
selected_videos = pd.read_csv('/content/drive/MyDrive/accident_files/smaller_sample_videos.csv')

# First split: 70% train, 30% temp (for validation and test)
train_df, temp_df = train_test_split(selected_videos, test_size=0.3, shuffle=True, random_state=42, stratify=selected_videos['weather'])
train_df = train_df.copy()

extracted_video_count = 0
total_frames_extracted = 0

# Frame extraction loop
for index, row in train_df.iterrows():
    relative_rgb_path = row['rgb_path'] # e.g., videos/head-on/Town05_head-on_night_34.mp4
    full_video_path = os.path.join(base_video_dataset_path, relative_rgb_path)

    # Get the base filename without path or extension
    # e.g., 'videos/head-on/Town05_head-on_night_34.mp4' -> 'Town05_head-on_night_34'
    base_video_name = os.path.splitext(os.path.basename(relative_rgb_path))[0]
    filename_prefix = base_video_name

    cap = cv2.VideoCapture(full_video_path)

    if not cap.isOpened():
        print(f"    Error: Could not open video {full_video_path}")
        continue

    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break # End of video

        # Name file as videofilename_frame_xxxx.jpg
        frame_filename = f"{filename_prefix}_frame_{frame_count:04d}.jpg" # Using 4 digits for frame number
        frame_output_path = os.path.join(output_frames_dir, frame_filename)

        cv2.imwrite(frame_output_path, frame)
        frame_count += 1

    cap.release()
    print(f"    Extracted {frame_count} frames from {relative_rgb_path}")
    extracted_video_count += 1
    total_frames_extracted += frame_count

print(f"\nCompleted frame extraction for {extracted_video_count} selected videos. Total frames: {total_frames_extracted}.")
```
